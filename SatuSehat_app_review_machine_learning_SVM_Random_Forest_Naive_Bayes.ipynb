# Filtering (Stopword Removal)

**Filtering** adalah tahap mengambil kata-kata penting dari hasil token dengan menggunakan algoritma stoplist (membuang kata kurang penting) atau wordlist (menyimpan kata penting).

**Stopword** adalah kata umum yang biasanya muncul dalam jumlah besar dan dianggap tidak memiliki makna. Contoh stopword dalam bahasa Indonesia adalah “yang”, “dan”, “di”, “dari”, dll. Makna di balik penggunaan stopword yaitu dengan menghapus kata-kata yang memiliki informasi rendah dari sebuah teks, kita dapat fokus pada kata-kata penting sebagai gantinya.

Contoh penggunaan filtering dapat kita temukan pada konteks mesin pencarian. Jika permintaan pencarian anda adalah “apa itu pengertian manajemen?” tentunya anda ingin sistem pencarian fokus pada memunculkan dokumen dengan topik tentang “pengertian manajemen” di atas dokumen dengan topik “apa itu”. Hal ini dapat dilakukan dengan mencegah kata dari daftar stopword dianalisa.

# Stemming sastrawi

Sastrawi Python adalah library python sederhana yang memungkinkan Anda untuk mereduksi kata-kata infleksi dalam Bahasa Indonesia (Bahasa Indonesia) ke bentuk dasarnya (stem). Ini adalah port Python dari proyek Sastrawi asli yang ditulis dalam PHP (kredit diberikan kepada penulis asli dan kontributor Sastrawi PHP). https://medium.com/@93Kryptonian/stemming-with-sastrawi-877cc40a37ad

Stemming adalah teknik pemrosesan bahasa alami yang menurunkan infleksi kata ke bentuk akarnya, sehingga membantu dalam pra-pemrosesan teks, kata, dan dokumen untuk normalisasi teks.

Menurut Wikipedia, infleksi adalah proses di mana sebuah kata dimodifikasi untuk mengkomunikasikan banyak kategori tata bahasa, termasuk tegang, kasus, suara, aspek, orang, jumlah, jenis kelamin, dan suasana hati. Jadi, meskipun sebuah kata mungkin ada dalam beberapa bentuk infleksi, memiliki beberapa bentuk infleksi di dalam teks yang sama menambah redundansi pada proses NLP.

Akibatnya, kami menggunakan stemming untuk mengurangi kata ke bentuk dasar atau stem, yang mungkin atau mungkin bukan kata yang sah dalam bahasa tersebut.

Misalnya, batang dari ketiga kata ini, koneksi, terhubung, menghubungkan, adalah "menghubungkan". Di sisi lain, akar masalah, masalah, dan masalah adalah "masalah", yang bukan kata yang dikenal.

# Penerapan TF-IDF
Term frequency-inverse document frequency adalah text vectorizer yang mengubah teks menjadi vektor yang dapat digunakan. Ini menggabungkan 2 konsep, Term Frequency (TF) dan Document Frequency (DF).

Frekuensi istilah adalah jumlah kemunculan istilah tertentu dalam dokumen. Frekuensi istilah menunjukkan seberapa penting istilah tertentu dalam sebuah dokumen. Frekuensi istilah mewakili setiap teks dari data sebagai matriks yang barisnya adalah jumlah dokumen dan kolomnya adalah jumlah istilah yang berbeda di seluruh dokumen.

Frekuensi dokumen adalah jumlah dokumen yang mengandung istilah tertentu. Frekuensi dokumen menunjukkan seberapa umum istilah tersebut.

Inverse document frequency (IDF) adalah bobot suatu term, hal ini bertujuan untuk mengurangi bobot suatu term jika kemunculan term tersebar di seluruh dokumen. IDF dapat dihitung sebagai berikut:
rumus idf.webp


Dimana idfᵢ adalah skor IDF untuk term i, dfᵢ adalah jumlah dokumen yang mengandung term i, dan n adalah jumlah total dokumen. Semakin tinggi DF suatu term, semakin rendah IDF untuk term tersebut. Ketika jumlah DF sama dengan n yang berarti istilah tersebut muncul di semua dokumen, IDF akan menjadi nol, karena log(1) adalah nol, jika ragu masukkan saja istilah ini ke dalam daftar stopword karena tidak menyediakan banyak informasi.

Skor TF-IDF seperti namanya hanyalah perkalian dari matriks frekuensi term dengan IDF-nya, dapat dihitung sebagai berikut:

idf.png

Dimana wᵢⱼ adalah skor TF-IDF untuk term i pada dokumen j, tfᵢⱼ adalah frekuensi term untuk term i pada dokumen j, dan idfᵢ adalah skor IDF untuk term i.

# HANDLING IMBALANCE
Dalam Machine Learning dan Data Science sering kita jumpai istilah yang disebut Distribusi Data Tidak Seimbang, umumnya terjadi ketika pengamatan di salah satu kelas jauh lebih tinggi atau lebih rendah dari kelas lainnya. Karena algoritma Machine Learning cenderung meningkatkan akurasi dengan mengurangi kesalahan, mereka tidak mempertimbangkan distribusi kelas.

Masalah ini lazim dalam contoh-contoh seperti Deteksi Penipuan, Deteksi Anomali, Pengenalan wajah, dll.

Teknik ML standar seperti Decision Tree dan Logistic Regression memiliki bias terhadap kelas mayoritas, dan cenderung mengabaikan kelas minoritas. Mereka cenderung hanya memprediksi kelas mayoritas, oleh karena itu, memiliki kesalahan klasifikasi yang besar dari kelas minoritas dibandingkan dengan kelas mayoritas.

Dengan kata yang lebih teknis, jika kita memiliki distribusi data yang tidak seimbang dalam kumpulan data kita, maka model kita menjadi lebih rentan terhadap kasus ketika kelas minoritas memiliki ingatan yang dapat diabaikan atau sangat rendah.
